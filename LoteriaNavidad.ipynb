```python
import json
import urllib.request
import argparse
import time
from datetime import datetime

```

```python
def print_sepLine(char, num):
    string = char * num
    print (string)
```

```python
def file2dictionary(file2read, split_char):
    """Read file and generate a dictionary"""
    d = {}
    with open(file2read) as f:
        for line in f:
            line = line.rstrip('\n')
            (key, val) = line.split(split_char)
            d[key] = val

    return(d)
```

```python
##########################################################
def consultar(n):
    #url_elpais = "https://api.elpais.com/ws/LoteriaNavidadPremiados?n=resumen" + n
    #respuesta = urllib.request.urlopen(url_otra)
    #contenido = respuesta.read()
    #datos = json.loads(co#ntenido.decode("utf8").replace("busqueda=", ""))
    
    ## https://pabloclementeperez.wordpress.com/2023/12/31/pesadilla-antes-de-navidad-con-la-app-de-la-loteria/
    url_otra = "https://api-loteria.pabloclementeperez.com/output/LoteriaNavidad.json"
    respuesta = urllib.request.urlopen(url_otra)

    # storing the JSON response  
    # from url in data 
    data_json = json.loads(respuesta.read()) 

    # print the json response 
    #print(data_json)
    
    premio=0
    try:
        premio = int(data_json[n])
        
    except KeyError:
        print()
    
    return premio
```

```python
##########################################################
def sorteo(dictionary_numeros):
    
    total_ganado = 0.0
    total_jugado = 0.0
    for num in dictionary_numeros.keys():
        info_num = dictionary_numeros[num].split("-")
        jugado=info_num[0]
        ident = info_num[1]
        
        ganado_decimo = consultar(str(int(num)))
        he_ganado = max(float(jugado) * ganado_decimo / 20, 0)
        total_ganado += he_ganado
        total_jugado += float(jugado)
        print_sepLine("*", 50)
        print(
            "Número: " + num + "\n",
            "  Jugado: " + str(jugado) + " euros" + "\n",
            "  Ganado: " + str(he_ganado) + " euros" + "\n",
            "  Ident: " + ident + "\n",
        )
        
    print("\n=============================")
    print("Total Jugado = " + "{0:10.2f}".format(total_jugado), "euros")
    print("Total Ganado = " + "{0:10.2f}".format(total_ganado), "euros")
    print("Saldo        = " + "{0:10.2f}".format(total_ganado - total_jugado), "euros")

    print()
    print()
```

```python
###############   
def print_time ():
    """Prints time stamp in human readable format: month/day/year, hour:minute:seconds."""
    now = datetime.now()
    date_time = now.strftime("%m/%d/%Y, %H:%M:%S")
    print ('\t' + date_time + '\n')
```
